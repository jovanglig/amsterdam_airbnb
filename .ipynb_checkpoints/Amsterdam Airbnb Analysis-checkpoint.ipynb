### Clean data function

Below we define our clean data function using our findings above, which returns our X and y variables for our model

```python
def clean_data(df):
    '''
    Clean data 
    
    df = dataframe
    '''
    # Dropping columns and rows with all missing values
    df = df.dropna(axis = 0, how = 'all')
    df = df.dropna(axis = 1, how = 'all')
    
    #Dropping ID fields by their common dtype INT64
    id_columns = df.select_dtypes(include=['int64']).columns[:3]
    df = df.drop(id_columns, axis=1)
    
    #Dropping unnecessary string fields in our data
    df = df.drop(['host_url','host_since','host_verifications', 'listing_url', 'host_about', 'latitude', 'longitude', 'host_location', 'neighborhood_overview', 'description', 'picture_url',
             'host_name', 'bathrooms_text','neighbourhood', 'host_picture_url', 'host_thumbnail_url', 'name', 'calendar_last_scraped'], axis = 1)
    
    #Making sure fields meant to be boolean are actually boolean
    vars_boolean = df.columns[np.sum(df.isin(['t', 'f'])) > 0].to_list()
    for var in vars_boolean:
        df[var] = df[var].map({'t' : True, 'f': False}, na_action = 'ignore')
    
                  
    #Getting the values from the amenities column and creating booleans
    #Creating the amenities_list from amenities
    amenities_list = list(df.amenities)
    #Cleaning the list
    for item in range(len(amenities_list)):
        amenities_list[item] = amenities_list[item].strip('][').split(', ')
    for word in range(len(amenities_list[item])):
        amenities_list[item][word] = amenities_list[item][word].replace('"', '') 
    #Removing list of lists and creating full list    
    amenities_list = [item for sublist in amenities_list for item in sublist]  
    #Making the list unique and removing empty string
    amenities_set = set(amenities_list)
    amenities_list = list(amenities_set)
    amenities_list.remove(amenities_list[amenities_list == ''])
    
    #Cleaning up the original amenities column
    for item in range(len(df.amenities)):
        df.amenities[item] = df.amenities[item].strip('][').split(', ')
    for word in range(len(df.amenities[item])):
        df.amenities[item][word] = df.amenities[item][word].replace('"', '')
    
    #Creating bools for each amenity item in the amenities column matching against the unique amenities_list
    for item in amenities_list:
        df[item] = df.apply(lambda amenities: item in amenities['amenities'], axis = 1)
    #Dropping the amenities column
    df = df.drop(['amenities'], axis = 1)
    
    #Dropping columns with more than 30% missing values
    to_drop = df.columns[np.sum(df.isnull())/df.shape[0] > .30]
    df = df.drop(to_drop, axis = 1)
    
    
    #Defining date fields (transforming date fields into something meaningfull):
    df['last_scraped'] = df['last_scraped'].astype('datetime64')
    date_vars = df[['first_review', 'last_review']]
    for var in date_vars:
        df[var] = df[var].astype('datetime64')
        df[var] = df['last_scraped'] - df[var]
        df[var] = df[var] / pd.to_timedelta(1, unit='D')
        df = df.rename(columns = {var: 'days_since_'+var })
    df = df.drop(['last_scraped'], axis = 1)
    
    #Making sure price is a float and creating a night price based on minimum length of stay:
    df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
    df['night_price'] = df['price'] / df['minimum_nights']
    df = df.drop(['price'], axis = 1)
    
    #Removing outliers that can influence the regression line
    df = df[df['night_price'] < 1200]
    
    
    # Filling numeric columns with the mean
    num_vars = df.select_dtypes(include=['float', 'int']).columns
    for var in num_vars:
        df[var].fillna((df[var].mean()), inplace=True)  
        
    #Filling categorical columns with the mode and creating dummies and dropping original columns
    cat_vars = df.select_dtypes(include=['object']).copy().columns
    for var in  cat_vars:
        df[var].fillna((df[var].mode()), inplace = True)
        df = pd.concat([df.drop(var, axis=1), pd.get_dummies(df[var], prefix=var, prefix_sep='_', drop_first=True)], axis=1)
    
    
    #Remove highly correlated variables to minimize multicollinearity by using a correlation matrix
    corr_matrix = df.corr().abs()
    #Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    high_corr = [column for column in upper.columns if any(upper[column] > 0.95)]
    df = df.drop(high_corr, axis=1)
    
    
    
    y = df['night_price']
    df = df.drop(['night_price'], axis = 1)
    X = df 
    return X, y


 
```

## Data Modeling

### Feature selection

With more than 200 attributes our model seems rather complex. To improve prediction accuracy, I will perform feature selection using the following methods:

1. Covariance
2. Mutual Information Gain

Mutual information is calculated between two variables and measures the reduction in uncertainty for one variable given a known value of the other variable.

```python
cor = df.corr()
cor_target = abs(cor["night_price"])
cor_target.sort_values(ascending = False)[:11]
```

We can see above that the variables we have are not much correlated with the target variable unfortunately. Let's have a look at a correlation matrix where we apply a threshold of 10% correlation.

```python
plt.figure(figsize=(12,10))
relevant_features = cor_target[cor_target>0.1]
sns.heatmap(df[relevant_features.index].corr(), annot=True, cmap=plt.cm.Reds)
plt.show()
```

Here I am defining the feature selection function with scoring function equal to mutual information gain

```python
# feature selection
def select_features(X_train, y_train, X_test, k):
    '''
    Selects features based on mutual info regression
    Returns the following:
    training set with selected features using mutual info regression
    test set with selected features using mutual info regression
    feature selection object
    
    X_train = training set predictors
    y_train = training set target
    X_test = test set predictos
    k = n of features
    
    '''
    # configure to select a subset of features
    fs = SelectKBest(score_func=mutual_info_regression, k=k)
    # learn relationship from training data
    fs.fit(X_train, y_train)
    # transform train input data
    X_train_fs = fs.transform(X_train)
    # transform test input data
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

### Defining linear model and feature selection

Cleaning the data using our defined function above, splitting the dataset into train and test set (using 30% as test data), doing feature selection based on mutual information regression. Instantiating the model, fitting the model on the new dataset with selected features. Finally predicting and scoring the results using R2 as performance measure

```python
def clean_fit_linear_mod(df, test_size, rand_state, k):
    '''
    Cleans data and fits a linear model
    df = dataframe
    test_size = holdout data as decimal percentage
    rand_state = seed
    k = n of features
    '''
    #Clean the data and produce predictors (X) and label(y)
    X, y = clean_data(df)
    
    #Split into train and test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=rand_state)
    #Feature selection
    X_train_fs, X_test_fs, fs = select_features(X_train, y_train, X_test, k=k)
    
    #Instantiation
    lm_model = LinearRegression(normalize=True) 
    #Fitting model on selected features
    lm_model.fit(X_train_fs, y_train) 

    #Prediction
    y_test_preds = lm_model.predict(X_test_fs)
    y_train_preds = lm_model.predict(X_train_fs)
    
    #Scoring
    test_score = r2_score(y_test, y_test_preds)
    train_score = r2_score(y_train, y_train_preds)

    return test_score, train_score, lm_model, X_train_fs,X_train, X_test_fs, y_train, y_test, y_test_preds, fs


```

### Model with features that have a correlation of more than 10% with the price

Below is the clean data function with an added section of removing variables which do not have a 10% correlation with the target variable.

```python
def clean_data(df):
    '''
    Clean data 
    
    df = dataframe
    '''
        
    # Dropping columns and rows with all missing values
    df = df.dropna(axis = 0, how = 'all')
    df = df.dropna(axis = 1, how = 'all')
    
    #Dropping ID fields by their common dtype INT64
    id_columns = df.select_dtypes(include=['int64']).columns[:3]
    df = df.drop(id_columns, axis=1)
    
    #Dropping unnecessary string fields in our data
    df = df.drop(['host_url','host_since','host_verifications', 'listing_url', 'host_about', 'latitude', 'longitude', 'host_location', 'neighborhood_overview', 'description', 'picture_url',
             'host_name', 'bathrooms_text','neighbourhood', 'host_picture_url', 'host_thumbnail_url', 'name', 'calendar_last_scraped'], axis = 1)
    
    #Making sure fields meant to be boolean are actually boolean
    vars_boolean = df.columns[np.sum(df.isin(['t', 'f'])) > 0].to_list()
    for var in vars_boolean:
        df[var] = df[var].map({'t' : True, 'f': False}, na_action = 'ignore')
    
    #Getting the values from the amenities column and creating booleans
    #Creating the amenities_list from amenities
    amenities_list = list(df.amenities)
    #Cleaning the list
    for item in range(len(amenities_list)):
        amenities_list[item] = amenities_list[item].strip('][').split(', ')
    for word in range(len(amenities_list[item])):
        amenities_list[item][word] = amenities_list[item][word].replace('"', '') 
    #Removing list of lists and creating full list    
    amenities_list = [item for sublist in amenities_list for item in sublist]  
    #Making the list unique and removing empty string
    amenities_set = set(amenities_list)
    amenities_list = list(amenities_set)
    amenities_list.remove(amenities_list[amenities_list == ''])
    
    #Cleaning up the original amenities column
    for item in range(len(df.amenities)):
        df.amenities[item] = df.amenities[item].strip('][').split(', ')
    for word in range(len(df.amenities[item])):
        df.amenities[item][word] = df.amenities[item][word].replace('"', '')
    
    #Creating bools for each amenity item in the amenities column matching against the unique amenities_list
    for item in amenities_list:
        df[item] = df.apply(lambda amenities: item in amenities['amenities'], axis = 1)
    #Dropping the amenities column
    df = df.drop(['amenities'], axis = 1)
    
    #Dropping columns with more than 30% missing values
    to_drop = df.columns[np.sum(df.isnull())/df.shape[0] > .30]
    df = df.drop(to_drop, axis = 1)
    
    
    #Defining date fields (transforming date fields into something meaningfull):
    df['last_scraped'] = df['last_scraped'].astype('datetime64')
    date_vars = df[['first_review', 'last_review']]
    for var in date_vars:
        df[var] = df[var].astype('datetime64')
        df[var] = df['last_scraped'] - df[var]
        df[var] = df[var] / pd.to_timedelta(1, unit='D')
        df = df.rename(columns = {var: 'days_since_'+var })
    df = df.drop(['last_scraped'], axis = 1)
    
    #Making sure price is a float and creating a night price based on minimum length of stay:
    df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
    df['night_price'] = df['price'] / df['minimum_nights']
    df = df.drop(['price'], axis = 1)
    
    #Removing outliers that can influence the regression line
    df = df[df['night_price'] < 1200]
    
    
    # Filling numeric columns with the mean
    num_vars = df.select_dtypes(include=['float', 'int']).columns
    for var in num_vars:
        df[var].fillna((df[var].mean()), inplace=True)  
        
    #Filling categorical columns with the mode and creating dummies and dropping original columns
    cat_vars = df.select_dtypes(include=['object']).copy().columns
    for var in  cat_vars:
        df[var].fillna((df[var].mode()), inplace = True)
        df = pd.concat([df.drop(var, axis=1), pd.get_dummies(df[var], prefix=var, prefix_sep='_', drop_first=True)], axis=1)
    
    
    #Remove highly correlated variables to minimize multicollinearity by using a correlation matrix
    corr_matrix = df.corr().abs()
    #Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    high_corr = [column for column in upper.columns if any(upper[column] > 0.95)]
    df = df.drop(high_corr, axis=1)
    
    #Keep only correlated variables              
    corr_matrix = df.corr().abs()
    cor_target = corr_matrix['night_price']
    relevant_features = cor_target[cor_target>0.1]
    df = df[list(relevant_features.index)]
    
    
    y = df['night_price']
    df = df.drop(['night_price'], axis = 1)
    X = df 
    return X, y


 
```
